```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import math
```

```python
r = requests.get('https://www.npr.org/2022/12/15/1135802083/100-best-songs-2022-page-1')
r.status_code
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
songs = []
for i in soup.find_all('h3', attrs = {'class': 'edTag'}):
    songs.append(i.get_text())
songs
```

```python
len(songs) 
```

It can be seen that there is an extra value in the list which is not related to the songs and is just a website content. So, we will remove it.

```python
songs.remove('Next >')
```

```python
songs
```

```python
len(songs)
```

```python

# Convert a list to dictionary
 
def Convert(songs):
    res_dct = {songs[i]: songs[i + 1] for i in range(0, len(songs), 2)}
    return res_dct
         
print(Convert(songs))
```

```python
song_dict = Convert(songs)
```

```python
#Creating a dataframe from the doctionary
df = pd.DataFrame(song_dict.items(), columns=['artist', 'song'])
```
